```python
import string 
import re
```

```python
def remove_p(tex):
    temp = ''.join([char for char in tex if char not in string.punctuation])
    return temp

data['text'] = data['text'].apply(lambda x: remove_p(x))
```

```python
def tocken(txt):
    tockens = re.split('\W+',txt)
    return tockens

data['text'] = data['text'].apply(lambda x: tocken(x.lower()))
```

```python
data.head()
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_s(txt):
    temp = [word for word in txt if word not in stopwords]
    return temp

data['text'] = data['text'].apply(lambda x : remove_s(x))
```

```python
data.head()
```

```python
wn = nltk.WordNetLemmatizer()
```

```python
def lem(txt):
    temp = [wn.lemmatize(word) for word in txt]
    return temp

data['text'] = data['text'].apply(lambda x : lem(x))
```
